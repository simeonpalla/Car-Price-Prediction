```python
# sepereate x and y
df_attr=df_train.iloc[:,0:16]
df_y=df_train[['Price']]
```

```python
#Remove one of Age of Car or Manufacturing year
df_attr=df_attr.drop(columns='Age of car',axis=1)
df_attr.corr()
```

```python
#imputation and missing value  replacement
df_attr['Location'].value_counts()
```

```python
df_attr['Owner Type'] = df_attr['Owner Type'].replace({'First':1,'Second':2,'Third':3,'Fourth & Above':4})
df_attr['transmission']=df_attr['transmission'].replace({'man':0,'auto':1})
df_attr['fuel_type']=df_attr['fuel_type'].replace({'petrol':0,'diesel':1})
df_attr['door_count']=df_attr['door_count'].replace({'None': 4})
df_attr['seat_count']=df_attr['seat_count'].replace({'None': 5})
df_attr['Maker']=df_attr['Maker'].replace({'skoda':1,'toyota':2,'audi':3,'bmw':4,'nissan':5,'hyundai':6,'fiat':7,'maserati':8})
df_attr['model']=df_attr['model'].replace({ 'octavia':1
,'superb':2
,'yaris' :3
,'qashqai' :4
,'x3' :5
,'x1' :6
,'i30' :7
,'q5' :8
,'x5' :9
,'yeti' :10
,'panda' :11
,'q3' :12
,'coupe' :13
,'micra' :14
,'auris' :15
,'avensis' :16
,'aygo' :17
,'rapid' :18
,'roomster': 19
,'q7' :20
,'citigo':21
,'juke' :22
,'tt' :23})
df_attr['Location']=df_attr['Location'].replace({ 'Coimbatore':1,
'Kochi':2,     
'Bangalore':3, 
'Jaipur':4,    
'Kolkata':5,   
'Pune':6,      
'Mumbai':7,    
'Chennai':8,   
'Delhi':9,     
'Hyderabad':10,
'Ahmedabad':11,
})
```

```python
df_attr['Maker'].value_counts()
```

```python

df_attr.isnull().sum()
df_attr[['seat_count']].value_counts()
df_attr['seat_count']=df_attr['seat_count'].apply(int)
df_attr['door_count']=df_attr['door_count'].apply(int)
```

```python

df_attr[['door_count']].value_counts()
```

```python
df_attr.info()
```

```python
#MIssing value 
df_attr.isnull().sum()
#Lets to univariate/bivariate analyses analyses of Distance/engine_power/body_type
df_attr=df_attr.drop(columns='body_type',axis=1)# we can drop the column body_type as it is mosly null
df_attr.describe()


```

```python
#Distance
df_attr.columns
```

```python
sns.boxplot(x=df_attr['Distance '])
#df_attr['Distance '].describe()
df_attr['Distance ']=df_attr['Distance '].fillna(df_attr['Distance '].median()).apply(int)
df_attr.info()
df_attr['engine_power']=df_attr['engine_power'].fillna(df_attr['engine_power'].median()).apply(int)
df_attr.info()
```

```python
df_attr
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(df_attr, df_y, test_size=0.30, random_state=1)

regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
regression_model.score(X_test, y_test)
```

```python
df_test=pd.read_csv("Test_Data_Set.csv")
```

```python
from sklearn.preprocessing import PolynomialFeatures


poly = PolynomialFeatures(degree=2, interaction_only=True)
X_train_ = poly.fit_transform(X_train)

X_test_ = poly.fit_transform(X_test)

poly_clf = LinearRegression()

poly_clf.fit(X_train_, y_train)

y_pred = poly_clf.predict(X_test_)

print(poly_clf.score(X_test_, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

model=DecisionTreeRegressor(criterion='squared_error')
model.fit(X_train, y_train)
model.score(X_train, y_train)
model.score(X_test, y_test)  # performance on test data

predictions=model.predict(df_attr)
mse = mean_squared_error(df_y, predictions)
print(f'Mean Squared Error: {mse}')
```

```python

from sklearn.metrics import mean_squared_error,r2_score
from sklearn.ensemble import RandomForestRegressor
# Fitting Random Forest Regression to the dataset
regressor = RandomForestRegressor(n_estimators=24, random_state=0, oob_score=True)
 
# Fit the regressor with x and y data
regressor.fit(df_attr, df_y)
oob_score = regressor.oob_score_
print(f'Out-of-Bag Score: {oob_score}')
 
predictions = regressor.predict(df_attr)
# Evaluating the model
mse = mean_squared_error(df_y, predictions)
print(f'Mean Squared Error: {mse}')

r2 = r2_score(df_y, predictions)
print(f'R-squared: {r2}')
# Making predictions on the same data or new data
#predictions = regressor.predict(x)
 
#grid_search = GridSearchCV(regressor, param_grid, cv=5,
                     #  scoring='neg_mean_squared_error')
#grid_search.fit(housing_prepared, housing_labels)    
```
